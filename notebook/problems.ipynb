## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np

i = 0
outcomes = 0
experiments = 1000
while i < experiments:
    die1 = np.random.randint(1, 7)
    die2 = np.random.randint(1, 7)
    total_score = die1 + die2
    if total_score % 2 == 0 or total_score > 7:
        outcomes += 1
    i += 1
probability = outcomes/experiments
print("The probability that total score sum even or greater than 7 is : ",round(probability*100,2))
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"


take_3white_2red = 0
take_all_colors = 0

for i in range(1000):
    drew_balls = []
    for j in range(5):
        drew_balls.append(ball_box[np.random.randint(0,60)])
    
    drew_balls = np.array(drew_balls)
    white_balls = sum(drew_balls=='White')
    red_balls = sum(drew_balls=='Red')
    green_balls = sum(drew_balls=='Green')

    if white_balls == 3 and red_balls == 2:
        take_3white_2red +=1
    if white_balls == 5 or red_balls == 5 or green_balls == 5:
        take_all_colors +=1

probability_take_3white_2red = round((take_3white_2red / 1000) * 100,2)
probability_take_all_colors = round((take_all_colors / 1000)*100,2)
print (probability_take_3white_2red)
print (probability_take_all_colors)
```
